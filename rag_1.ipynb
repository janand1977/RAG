```python
import os
from openai import OpenAI

# Initialize the standard OpenAI client, but route it to Groq
client = OpenAI(
    base_url="https://api.groq.com/openai/v1", # Points to Groq's endpoint
    api_key=os.environ.get("GROQ_API_KEY")      # Uses your Groq key
)

response = client.responses.create(
    model="openai/gpt-oss-20b", # An open model running natively on Groq hardware
    input="Draft a polite email asking for a deadline extension."
)
print(response.output_text)
```
