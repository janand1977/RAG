```python
import os
import yfinance as yf
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_groq import ChatGroq


model= ChatGroq(model="openai/gpt-oss-20b")

# 2. Update your template to accept the LIVE metrics fetched from the API
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a stock analyzer for Indian markets. Return only the final metric values clearly—no chatty descriptions."),
    ("human", "Format these live metrics for {stock_name} (Ticker: {ticker}): P/E Ratio: {pe_ratio}, P/B Ratio: {pb_ratio}.")
])



output_parser = StrOutputParser()
chain = prompt_template | model | output_parser

ticker_symbol = "RELIANCE.NS"
stock = yf.Ticker(ticker_symbol)
stock_info = stock.info

# Extract the live metrics safely (defaults to "N/A" if not found)
live_pe = stock_info.get("trailingPE", "N/A")
live_pb = stock_info.get("priceToBook", "N/A")
company_name = stock_info.get("longName", "Indian Stock")

result = chain.invoke({"stockName": "ITC"})
print(result)
```
