```python
def binarySearch(nums: list[int], k: int) -> int:
    s=0
    e= len(nums)-1

    while(s<=e):
        mid = s + (e-s)//2
    
        if(nums[mid] == k):
            return mid
        elif(nums[mid] > k):
            e=mid-1
        else:
            s=mid+1

    return -1


lst = [1,2,3,4,5,6,7,8,9]
print(binarySearch(lst,7))

```
