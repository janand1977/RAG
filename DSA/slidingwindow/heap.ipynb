```python
import heapq

data = [4,2,1,6,7]

heapq.heapify(data)

while(len(data)):
    smallest = heapq.heappop(data)
    print(smallest)



data = [4,2,1,6,7]
data= [-x for x in data]

heapq.heapify(data)
print("=======================================")
while(len(data)):
    largest = -heapq.heappop(data)
    print(largest)
```
