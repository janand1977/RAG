```python
from dataclasses import dataclass


@dataclass(frozen=True)
class Emp:
   id:float  # here this class acts like instance variable and not class variable
   name:str

a = Emp(100,'anand')
b = Emp(101, 'jp')
c = Emp(102, 'jass')

mySet: set[Emp] ={a,a,b}
print(mySet)
mySecondSet={b,c}




print(mySecondSet.difference(mySet))


#  simple set check

x = 5
if a in mySet:
   print(f"mySet contains a: {x}")


if a in mySecondSet:
   print(f"mySet contains b: {x}")


# filter

filtered = {item for item in mySet if item.id > 100 }
print(filtered)


# iterate through a set
for item in mySet:
   print(item)

```

```python
dic: dict[Emp,int] = {}
dic[a] = 1
dic[b] =2
dic[c]=3


print(dic[b])

if b in dic:
    print("b is present")

print(len(dic))

dic.pop(b)
dic.pop(a)
status = dic.pop(a,None)
print(status)

dic.update({a:55}) # update just adds another item 
print(dic[a])

# iterate through a dictionary
for key,value in dic.items():
    print(key)
```

```python
# uusing dictionary
newdict:dict[Emp,int] ={}
cmpdict:dict[Emp,int]={}

empA= Emp("abba",5)
empB= Emp("abba",8)

newdict[empA] = 1
cmpdict[empB] = 1

print(newdict==cmpdict)


xx = []
xx.append(empB)
xx.append(empA)



uu = []
uu.append(empA)
uu.append(empB)

print(xx==uu)




```
